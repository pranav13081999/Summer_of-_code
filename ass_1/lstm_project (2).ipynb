```python
import numpy as np

import pandas as pd
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
text = open('Data.txt',encoding ="utf-8").read().lower()
tokenizer.fit_on_texts([text])
```

```python
len(tokenizer.word_index)
```

```python
tokenizer.word_index
```

```python
input_sequences = []
for sentence in text.split('.'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=1233)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense ,Dropout
```

```python
model = Sequential()
model.add(Embedding(1233, 400, input_length=83))
model.add(LSTM(3000))

model.add(Dropout(0.3))

model.add(Dense(1233, activation='softmax',kernel_regularizer=tf.keras.regularizers.l2(0.01)))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
import time

start_time = time.time()
model.fit(X,y,epochs=100)


end_time = time.time()
elapsed_time = end_time - start_time

print("Training time:", elapsed_time, "seconds")
```

```python
model.save('next_word_predictor1.h5')
```
